# grf package

```r
library("grf")
source("src/simulation/dgp.R")
source("src/plots/plot_effect_surface.R")
options(repr.plot.height=10, repr.plot.width=18)
```

```r
data_train <- dgp("complex", n=2000, d=2)

data_test <- rbind(
    c(0.1, 0.1), c(0.1, 0.8), c(0.8, 0.1), c(0.8, 0.8)
)
plot_true_effect("complex")
```

```r
cforest = grf::causal_forest(
    X=data_train[["X"]], Y=data_train[["Y"]], W=data_train[["W"]],
    num.trees=5000, tune.parameters="all", honesty=TRUE,
    num.threads=2, seed=0
)
```

```r
treatment_effect <- predict(cforest, data_test)[["predictions"]]
treatment_effect
```
