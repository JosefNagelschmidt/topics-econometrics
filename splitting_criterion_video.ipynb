```r
show_video <- function(filename, mimetype) {
    library(IRdisplay)
    library(base64enc)

    data = base64encode(filename, 'raw')

    display_html(paste0('<video controls src="data:',
         mimetype, ';base64,', data, '">'))
}
```

```r
show_video("img/splitting_criterion_cut.mp4", "video/mp4")
```
